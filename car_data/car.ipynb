```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the data from the CSV file
data = pd.read_csv('car_data.csv')

# Extract the Year and Selling_Price values
year = data['Year']
selling_price = data['Selling_Price']

# Create a line chart
plt.plot(year, selling_price, marker='o')

# Add title and labels
plt.title("Year vs. Selling Price")
plt.xlabel("Year")
plt.ylabel("Selling Price")

# Show the chart
plt.show()
```

```python
# Count the occurrences of each fuel type
fuel_type_counts = data['Fuel_Type'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(fuel_type_counts, labels=fuel_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Fuel Types')
plt.show()
```

```python
# Create a bar graph for fuel type counts
plt.figure(figsize=(10, 6))
fuel_type_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Count of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Count the occurrences of each seller type
seller_type_counts = data['Seller_Type'].value_counts()

# Create a bar chart for seller type counts
plt.figure(figsize=(10, 6))
seller_type_counts.plot(kind='bar', color=['blue', 'orange'])
plt.title('Count of Cars by Seller Type')
plt.xlabel('Seller Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Count the occurrences of each transmission type
transmission_counts = data['Transmission'].value_counts()

# Print the counts
print(transmission_counts)

# Create a bar chart for transmission type counts
plt.figure(figsize=(10, 6))
transmission_counts.plot(kind='bar', color=['blue', 'orange'])
plt.title('Count of Cars by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
